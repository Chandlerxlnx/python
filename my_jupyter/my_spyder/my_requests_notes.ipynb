# Response
## attributes of response

```python
import requests

response = requests.get('http://www.jianshu.com')
print(type(response.status_code),response.status_code)
print(type(response.headers),response.headers)
print(type(response.cookies),response.cookies)
print(type(response.url),response.url)
print(type(response.history),response.history)
```

## judgement of status code

```python
import requests

response = requests.get('http://www.jianshu.com')
print(response.status_code)
exit() if not response.status_code == requests.codes.ok else print('Request Successfully')
```

```python
import requests

response = requests.get('http://www.jianshu.com')
exit() if not response.status_code == 200 else print('Request Successfully')
```

## get cookies

```python
import requests

response = requests.get('https://www.baidu.com/')
print(response.cookies)
for key,value in response.cookies.items():
    print(key+'='+value)
```

## use the cookies to stick the session (会话维持)
use cookies to keep login state

```python
import requests

requests.get('http://httpbin.org/cookies/set/number/123456789')
response=requests.get('http://httpbin.org/cookies')
print(response.text)
```

```python
import requests

s = requests.Session()
s.get('http://httpbin.org/cookies/set/number/123456789')
response=s.get('http://httpbin.org/cookies')
print(response.text)
```

```python
## seting proxy
import requests

proxies = {
    "http":"http://127.0.0.1:9743",
    "https":"https://127.0.0.1:9743"
}
response =requests.get("https://www.taobao.com",proxies=proxies)
print(response.status_code)
```

```python
## seting proxy
import requests

proxies = {
    "http":"http://user.passwd@127.0.0.1:9743/",
    "https":"https://user.passwd@127.0.0.1:9743/",
}

response =requests.get("https://www.taobao.com",proxies=proxies)
print(response.status_code)
```

```python
## time out
import requests



#response =requests.get("https://www.taobao.com",timeout =1)

try:
    r =requests.get("http://httpbin.org/get",timeout =1)
    print(response.status_code)
except ReadTimeout:
    print(r.status_code)
```
